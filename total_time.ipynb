```python
import pandas as pd
```

```python
lap_times = pd.read_csv('lap_times.csv')
race_841 = lap_times.query('raceId == 841')
```

```python
race_841 = race_841.groupby('driverId', as_index = False).aggregate({'milliseconds' : 'sum'}).\
rename(columns = {'milliseconds' :'total_milliseconds'})
```

```python
race_841['milliseconds'] = race_841.total_milliseconds % 1000
race_841['minutes'] = (race_841.total_milliseconds // 1000) // 60
race_841['seconds'] = (race_841.total_milliseconds // 1000) % 60
race_841['hours'] = (race_841.minutes) // 60
race_841['minutes'] = race_841['minutes']  - 60 * race_841['hours']
```

```python
race_841['total_time'] = 0
for i in range(len(race_841)):
    race_841.iloc[i,6] = str(race_841.iloc[i,5]) + ":" + str(race_841.iloc[i,3]) + ":" \
    + str(race_841.iloc[i,4]) + "." + str(race_841.iloc[i,2])
```
