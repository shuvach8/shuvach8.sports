```python
import pandas as pd
lap_times = pd.read_csv('lap_times.csv')
race_841 = lap_times.query('raceId == 841')
```

```python
race_841.query('lap == 57')
```

```python
lap_positions = pd.DataFrame({'lap': [i for i in range(1,59)]}, \
                             columns = ['lap', 'first_driver', 'second_driver', 'third_driver'])
```

```python
for i in range(1,59): 
    k = i
    lap_positions.iloc[i-1, 0] = i
    lap_positions.iloc[i-1, 1] = race_841.query('lap == @k').iloc[0,1]
    lap_positions.iloc[i-1, 2] = race_841.query('lap == @k').iloc[1,1]
    lap_positions.iloc[i-1, 3] = race_841.query('lap == @k').iloc[2,1]
```
