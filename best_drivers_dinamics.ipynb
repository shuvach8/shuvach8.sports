```python
import pandas as pd
lap_times = pd.read_csv('lap_times.csv')
race_841 = lap_times.query('raceId == 841')
```

```python
race_841.query('lap == 58').sort_values(['position'], ascending = True)
```

```python
best_drivers_dinamics = pd.DataFrame({'lap' : [i for i in range(1,59)]}, columns = ['lap', 'driverId_20',\
                                        'driverId_1', 'driverId_808', 'driverId_4', \
                                       'driverId_17', 'driverId_18', 'driverId_815', 'driverId_155', 'driverId_13'  ])
```

```python
n = 0
df = race_841.query('lap == 58').sort_values(['position'], ascending = True)

for i in (df.driverId.unique()):
    n += 1
    for p in range(58):
        best_drivers_dinamics.iloc[p, n] = race_841.query('driverId == @i').iloc[p,3]
```
