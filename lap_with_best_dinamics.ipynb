```python
import pandas as pd
```

```python
lap_times = pd.read_csv('lap_times.csv')
race_841 = lap_times.query('raceId == 841')
```

```python
ans_dict = { 'driverId' : [], 'best_lap' : [], 'best_dinamics' : [], } 
for n in (race_841.driverId.unique()):
    driver_results = race_841[race_841.driverId == n]
    ans_dict['driverId'].append(n)
    ans_dict['best_lap'].append(0)
    ans_dict['best_dinamics'].append(0)
    for lap in range(driver_results.lap.min() + 1, driver_results.lap.max()):
        dinamics = driver_results[driver_results.lap == lap].iloc[0]['position'] \
        - driver_results[driver_results.lap == lap - 1].iloc[0]['position']
        if dinamics > ans_dict['best_dinamics'][-1]:
            ans_dict['best_dinamics'][-1] = dinamics
            ans_dict['best_lap'][-1] = lap

ans = pd.DataFrame(ans_dict)
ans.style.hide_index()
    
```
